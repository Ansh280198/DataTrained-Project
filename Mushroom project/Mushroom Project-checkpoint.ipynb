```python
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import chi2
```

```python
#Instantiate selectpercentile and fit(feature,label)
X=data.drop(columns=['class'],axis=1)
y=data['class']

Spercentile = SelectPercentile(score_func=chi2, percentile=80)
Spercentile.fit(X,y)
```

```python
# seperate the features to check p_values
cols = Spercentile.get_support(indices=True)
features = X.columns[cols]
```

```python
df_scores = pd.DataFrame({'features':X.columns, 'Chi2score':Spercentile.scores_, 'pvalue':Spercentile.pvalues_,})
df_scores.sort_values(by='Chi2score',ascending=True)
```

```python
print(cols)
print(features)
```

```python
X=data[features]
y=data['class']
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
maxAcc = 0 #maximum accuracy
maxRs = 0 #best random state value for which max accuracy is achieved 
for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.20,random_state=i)
    LR = LogisticRegression()
    LR.fit(x_train,y_train) #fitting the data will train your model
    predlr = LR.predict(x_test) #this is the predicted target variable
    acc = accuracy_score(y_test,predlr)
    if acc>maxAcc:
        maxAcc = acc
        maxRs = i
print("Best accuracy is",maxAcc,"on random state",maxRs)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=82)
```

# Logisitic Regression 

```python
Model = LogisticRegression()
Model.fit(x_train,y_train)
pre_1 = Model.predict(x_test)
acc1 = accuracy_score(y_test,pre_1)
print(acc1)
```

```python
print(classification_report(y_test,pre_1))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
pred_dt = dt.predict(x_test)
acc2 = accuracy_score(y_test,pred_dt)
print(acc2)
print(classification_report(y_test,pred_dt))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
pred_rf = rf.predict(x_test)
acc3 = accuracy_score(y_test,pred_rf)
print(acc3)
print(classification_report(y_test,pred_rf))
```

# Gradient Boosting classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)
pred_gb = gb.predict(x_test)
acc4 = accuracy_score(y_test,pred_rf)
print(acc4)
print(classification_report(y_test,pred_gb))
```

# For hypertuning we will use grid search CV

```python
from sklearn.model_selection import GridSearchCV

#creating parameter list to pass in Gridsearchcv
parameter ={'learning_rate': np.arange(0.2,0.5,0.2),
           'max_depth':range(4,8),
            'min_samples_split':range(2,6)
           }
```
